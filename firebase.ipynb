```python
import pandas as pd
```

```python
df = pd.read_csv('./data/dataset_completo.csv')
```

```python
df = df[df['Cuatrimestre'].isin(['Primer Cuat.', 'Segundo Cuat.'])]

# Use apply instead of map to avoid Series within Series issue
df['Cuatrimestre_Id'] = df.apply(lambda row: '1Q'+str(row['Año']) if row['Cuatrimestre'] == 'Primer Cuat.' else '2Q'+str(row['Año']), axis=1)

df = df.drop(columns=['Cuatrimestre',"Año"])

```

```python
df["Horario"] = df["Horario"].apply(eval)
df = df.explode("Horario")
```

```python
info = {}

for _, data in df.iterrows():
    if data["Cuatrimestre_Id"] not in info:
        info[data["Cuatrimestre_Id"]] = {data["Codigo"]:{"name":data["Materia"],
                                         }}
    if data["Materia"] not in info[data["Cuatrimestre_Id"]]:
        info[data["Cuatrimestre_Id"]][data["Materia"]] = {}
    if data["Comision"] not in info[data["Cuatrimestre_Id"]][data["Materia"]]:
        info[data["Cuatrimestre_Id"]][data["Materia"]][data["Comision"]] = []
    

```
